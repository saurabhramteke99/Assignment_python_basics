Que_5. What is mutability in Python ?

Ans: In Python, mutability refers to whether or not an object can be changed after it is created. It determines whether an object can be modified in place or if a new object must be created to reflect the changes.

Que_6 Why are lists mutable, but tuples are immutable ?

Ans: Lists are mutable because they are designed for situations where the data needs to change (e.g., adding, removing, or modifying elements). Tuples are immutable to provide data integrity, ensuring their contents cannot be altered after creation. This makes tuples more efficient for fixed data and allows them to be used as dictionary keys, while lists are better for dynamic data.

Ans: In Python, the == and is operators are used for comparison, but they serve different purposes:

A) == (Equality Operator):

1)Compares values of two objects to check if they are equal.

2)It checks whether the contents or data of the objects are the same.

B)  is (Identity Operator):

1)Compares the memory addresses (identity) of two objects to check if they are the same object.

Que_9 What is type casting in Python ?

Ans: Type casting in Python refers to the process of converting one data type into another. It allows you to change the type of a value to suit your needs. There are two main types of type casting:

Implicit Casting (Automatic): Python automatically converts one data type to another when needed, usually when converting from a smaller type to a larger one.

Explicit Casting (Manual): You manually convert a value from one type to another using built-in functions like int(), float(), str(), etc.




Que_10 What is the difference between implicit and explicit type casting ?

Ans: A)Implicit Type Casting (Automatic):


1)Performed by Python automatically.

2)Happens when converting from a smaller data type to a larger data type (e.g., int to float).

3)No explicit action is required by the programmer.

4)Safe because no data is lost in the conversion.


B)Explicit Type Casting (Manual):


1)Performed by the programmer using functions like int(), float(), str(), etc.

2)Programmer manually converts one data type to another.

3)Can result in data loss or errors if not done correctly (e.g., converting a string to an integer).

4)Requires explicit function calls for conversion.

Que_12  How does the elif statement work?

Ans: The elif statement in Python stands for "else if" and is used to check multiple conditions in a sequence after an initial if statement. It allows you to test several conditions and execute the corresponding block of code for the first condition that is True. If none of the if or elif conditions are True, the else block (if present) will be executed.

How elif works:


1)Starts after an if statement: The elif statement follows an if block to check additional conditions if the if condition is False.

2)Evaluates conditions sequentially: If the if condition is False, the program moves to check the elif condition(s). The first True condition will execute its associated block of code.

3)Multiple elif statements: You can have multiple elif conditions to check several possibilities.

Ans: 1) for loop:

```python
#Que_14 . Write a program to reverse a string using a while loop.
def reversed_str(input_str):
  reversed_str = ""
  index = len(input_str) - 1
  while index >= 0:
    reversed_str += input_str[index]
    index -= 1
  return reversed_str
```

```python
reversed_str("saurabh")
```

```python
#Que_15  Write a Python program to calculate the factorial of a number provided by the user using a while loop

def factorial(num):
  fact = 1
  while num > 0:
    fact = fact * num
    num = num - 1
  return fact
```

```python
factorial(5)
```
